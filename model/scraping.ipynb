```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

base_url_links = []
for i in range(1, 2):
    base_url = f"https://www.batamnews.co.id/berita/{i}"
    base_url_links.append(base_url)

links = []
for url in base_url_links:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    # find all article links
    for a in soup.find_all("a", href=True):
        href = a["href"]
        # filter only news articles
        if "berita-" in href and href.endswith(".html"):
            # make full URL
            if href.startswith("http"):
                full_url = href
            else:
                full_url = "https://www.batamnews.co.id/" + href
            links.append(full_url)



links = list(set(links))

print("Article links have been obtained!")

titles = []
content = []
published_date = []
news_type = []
for news_links in links:
    response = requests.get(news_links)
    soup = BeautifulSoup(response.text, 'html.parser')

    get_news_type_div = soup.find('div', class_='post-date')
    get_news_type = get_news_type_div.find_all('a')
    news_type.append(get_news_type[1].text)

    get_titles = soup.find_all('h1')
    titles.append(get_titles[2].text)

    get_content = soup.find_all('p')
    article = ' '.join([p.get_text(strip=True) for p in get_content])
    content.append(article)

    get_published_date = soup.find_all('div', itemprop='datePublished')
    published_date.append(get_published_date[0].text)

print("Data has been obtained")

df = pd.DataFrame({
    'title' : titles,
    'news_type' : news_type,
    'content' : content,
    'published_date' : published_date,
    'links' : links
})

print("Data saved to csv!")
```

#### importing data

```python
#baca hasil crawling
import pandas as pd

df = pd.read_csv('crawling_batamnews.csv')
```

```python
#processing text: hapus kalimat berulang di setiap artikel
df['content'] = df['content'].str.replace(
    'Copyright  2014 -batamnews.co.idAll rights reservedDikelola dan Dikembangkan Oleh :PT. Batam Media Siber',
   '', regex=False
)

#buat kolom baru: title + content artikel
df['full_text'] = df['title'] + '' + df['content']
df['full_text']

df[['content', 'title', 'full_text']].head()
```

```python
import re

def clean_text(text):
    #lowercase
    text = text.lower()

    #remove any url
    text = re.sub(r'http\S+|www\S+', '', text)

    #remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    #remove numbers
    text = re.sub(r'\d+', '', text)

    #remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text

df['clean_content'] = df['content'].apply(clean_text)
df['clean_title'] = df['title'].apply(clean_text)
df['clean_full_text'] = df['full_text'].apply(clean_text)
df[['content', 'clean_content', 'title', 'clean_title', 'full_text', 'clean_full_text']].head(2)
```

```python
df['published_date'] = df['published_date'].str.extract(r'(\d{2} [A-Za-z]+ \d{4})')
df['published_date']
```

#### traditional text processing

```python
#tokenization
from nltk.tokenize import word_tokenize

def tokenization(text):
    tokens = word_tokenize(text)
    return tokens

df['tokenize_content'] = df['clean_content'].apply(tokenization)
df['tokenize_title'] = df['clean_title'].apply(tokenization)
df['tokenize_full_text'] = df['clean_full_text'].apply(tokenization)
df[['tokenize_content', 'tokenize_title', 'tokenize_full_text']].head(2)
```

```python
#stopword removal (dan, yang di, ke, the, is)
from nltk.corpus import stopwords

def stopword_removal(text):
    stop_words = set(stopwords.words('indonesian'))
    filtered_text = [w for w in text if w not in stop_words]
    return filtered_text

df['stopword_removal_content'] = df['tokenize_content'].apply(stopword_removal)
df['stopword_removal_title'] = df['tokenize_title'].apply(stopword_removal)
df['stopword_removal_full_text'] = df['tokenize_full_text'].apply(stopword_removal)
df[['stopword_removal_content', 'stopword_removal_title', 'stopword_removal_full_text']].head(2)
```

```python
#jumlah kata sebelum dan sesudah stopword removal
print(len(df['tokenize_content'][1]))
print(len(df['stopword_removal_content'][1]))
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

def stemming(text):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    return [stemmer.stem(word) for word in text]

df['stemming_content'] = df['stopword_removal_content'].apply(stemming)
df['stemming_title'] = df['stopword_removal_title'].apply(stemming)
df['stemming_full_text'] = df['stopword_removal_full_text'].apply(stemming)
df[['stemming_content', 'stemming_removal_title', 'stemming_removal_full_text']].head(2)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# tokenizer
tokenizer = AutoTokenizer.from_pretrained(
    "cahya/t5-base-indonesian-summarization-cased"
)

# model
model = AutoModelForSeq2SeqLM.from_pretrained(
    "cahya/t5-base-indonesian-summarization-cased"
)

# tokenize input
inputs = tokenizer(
    df['clean_content'][0],
    max_length=512,
    truncation=True,
    return_tensors='pt'
)

# generate summary
summary_ids = model.generate(
    inputs['input_ids'],
    max_length=80,
    min_length=20,
    num_beams=4
)

# decode summary
summary = tokenizer.decode(
    summary_ids[0],
    skip_special_tokens=True
)

print(summary)
```
